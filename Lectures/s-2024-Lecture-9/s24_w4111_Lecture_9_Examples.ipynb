Summaries, reports, computations, ... ... often have table-like result formats.


Consider customer revenue summaries.

````python
%%sql

use classicmodels;

with one as (
    select customerNumber, orderNumber, status,
           orderLineNumber, quantityOrdered, priceEach
    from
        orders join orderdetails using(orderNumber)
),
    two as (
        select
            customerNumber, orderNumber, sum(priceEach*one.quantityOrdered) as revenue
        from
            one
        group by customerNumber, orderNumber
    ),
    three as (
        select
            customerNumber, customerName, count(*) as noOfOrders, sum(revenue) as totalRevenue
        from
            customers join two using(customerNumber)
        group by customerNumber
    )
select * from three order by totalRevenue desc;
````

How do I get a table like structure from a list of nested documents and then do aggregation?


The ```unwind``` operator creates a new top-level document for each subdocument.


The ```project``` operator can pull fields up to the top from embedded documents.

````python

result = client['Classic_Models']['Customers'].aggregate(
    [
        {
            '$unwind': {
                'path': '$orders', 
                'preserveNullAndEmptyArrays': False
            }
        }, {
            '$unwind': {
                'path': '$orders.details', 
                'preserveNullAndEmptyArrays': False
            }
        }, {
            '$project': {
                '_id': 0, 
                'customerNumber': '$number', 
                'customerName': '$name', 
                'orderNumber': '$orders.number', 
                'orderStatus': '$orders.status', 
                'quantityOrdered': '$orders.details.quantityOrdered', 
                'priceEach': '$orders.details.priceEach', 
                'lineItemRevenue': {
                    '$multiply': [
                        '$orders.details.quantityOrdered', '$orders.details.priceEach'
                    ]
                }
            }
        }
    ]
)
````

````python
revenue_df = pandas.DataFrame(result)
revenue_df
````

There are the other obvious operators, e.g. ```groupBy, orderBy, ...```

````python
result = client['Classic_Models']['Customers'].aggregate([
    {
        '$unwind': {
            'path': '$orders', 
            'preserveNullAndEmptyArrays': False
        }
    }, {
        '$unwind': {
            'path': '$orders.details', 
            'preserveNullAndEmptyArrays': False
        }
    }, {
        '$project': {
            '_id': 0, 
            'customerNumber': '$number', 
            'customerName': '$name', 
            'orderNumber': '$orders.number', 
            'orderStatus': '$orders.status', 
            'quantityOrdered': '$orders.details.quantityOrdered', 
            'priceEach': '$orders.details.priceEach', 
            'lineItemRevenue': {
                '$multiply': [
                    '$orders.details.quantityOrdered', '$orders.details.priceEach'
                ]
            }
        }
    }, {
        '$group': {
            '_id': {
                'customerNumber': '$customerNumber', 
                'customerName': '$customerName'
            }, 
            'totalRevenue': {
                '$sum': '$lineItemRevenue'
            }, 
            'noOfOrders': {
                '$addToSet': '$orderNumber'
            }
        }
    }, {
        '$project': {
            'customerNumber': '$_id.customerNumber', 
            'customerName': '$_id.customerName', 
            '_id': 0, 
            'noOfOrders': {
                '$size': '$noOfOrders'
            }, 
            'totalRevenue': 1
        }
    }, {
        '$sort': {
            'totalRevenue': -1
        }
    }
])
````

````python
summary_df = pandas.DataFrame(list(result))
summary_df
````

````python
# Requires the PyMongo package.
# https://api.mongodb.com/python/current


result = client['Classic_Models']['Customers'].aggregate([
    {
        '$project': {
            '_id': 0, 
            'number': 1, 
            'name': 1, 
            'salesRepEmployeeNumber': 1
        }
    }, {
        '$lookup': {
            'from': 'Employees', 
            'localField': 'salesRepEmployeeNumber', 
            'foreignField': 'employeeNumber', 
            'as': 'salesRep'
        }
    }, {
        '$unwind': {
            'path': '$salesRep', 
            'preserveNullAndEmptyArrays': False
        }
    }, {
        '$project': {
            'number': '$number', 
            'name': '$name', 
            'employeeNumber': '$salesRep.employeeNumber', 
            'repLastName': '$salesRep.lastName', 
            'repFirstName': '$salesRep.firstName'
        }
    }
])
````

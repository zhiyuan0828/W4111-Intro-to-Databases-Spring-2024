- And even with the heuristic parser, it got a nickname wrong.<br><br>
- There are more sophisticated, parameterizable parsers.

````python
def move_some_rows(offset, limit):


    sql = f"select nconst, primaryName from name_basics limit {limit} offset {offset}"

    rows_to_move = %sql $sql

    rows_to_insert = []

    for r in rows_to_move:
        new_r = HumanName(r[1]).as_dict()
        new_r["nconst"] = r[0]

        for k,v in new_r.items():
            if new_r[k] == '':
                new_r[k] = None

        rows_to_insert.append(new_r)

    df = pandas.DataFrame(rows_to_insert)
    result = df.to_sql('name_basics_clean', con=engine, schema='s24_imdb_clean', if_exists='append', index=False)
    return result
    
````

````python
%sql drop table if exists s24_imdb_clean.name_basics_clean

limit = 10000
offset = 0
done = False
total = 0

while not done:
    count = move_some_rows(offset, limit)
    total += count
    print(f"Moved {count} rows from {offset} to {limit} for a total of {total} rows.")
    if count < limit:
        done = True
    else:
        offset = offset + limit

````

## Clean Up Multi-Value Attribute primaryProfession

- There are a lot of string manipulation functions in SQL, most of which do the obvious things.<br><br>
- Let's figure how many elements are in the comma list in ```primaryProfession.```

````python
%%sql

select
    length(primaryProfession) - length(replace(primaryProfession, ',', '')) as no_of_commas,
    primaryProfession, length(primaryProfession) primaryProfessionLength,
    replace(primaryProfession, ',', '') as no_commas_primaryProfession,
    length(replace(primaryProfession, ',', '')) as no_commas_primaryProfession_length

from
    name_basics
limit 20;
````

````python
%%sql

with one as (
select
        length(primaryProfession) - length(replace(primaryProfession, ',', '')) as no_of_commas,
    primaryProfession, length(primaryProfession) primaryProfessionLength,
    replace(primaryProfession, ',', '') as no_commas_primaryProfession,
    length(replace(primaryProfession, ',', '')) as no_commas_primaryProfession_length

from
    name_basics)
select max(no_of_commas) from one;
````

- Let's see if we can break it apart.

````python
%%sql

with one as (select
    nconst,
    primaryProfession,
    substr(primaryProfession, 1, locate(',', primaryProfession)-1) as p1,
    substr(primaryProfession, locate(',', primaryProfession)+1) as remainder
from
    name_basics),
    two as (
        select nconst, primaryProfession, p1,
               substr(remainder, 1, locate(',', remainder)-1) as p2,
               substr(remainder, locate(',', remainder)+1) as p3
        from
            one
    ),
    three as (
        select nconst,
               primaryProfession,
               if(p1='', NULL, p1) as p1,
               if(p2='', NULL, p2) as p2,
               if(p1='', NULL, p3) as p3
        from two
    )
select * from three limit 20;
````

- That is still not all that cool ... ... It kind of assumes that will never have more than 3.<br><br>
- A better design is ... ...

````python
%%sql

with one as (select
    nconst,
    primaryProfession,
    substr(primaryProfession, 1, locate(',', primaryProfession)-1) as p1,
    substr(primaryProfession, locate(',', primaryProfession)+1) as remainder
from
    name_basics),
    two as (
        select nconst, primaryProfession, p1,
               substr(remainder, 1, locate(',', remainder)-1) as p2,
               substr(remainder, locate(',', remainder)+1) as p3
        from
            one
    ),
    three as (
        select nconst,
               primaryProfession,
               if(p1='', NULL, p1) as p1,
               if(p2='', NULL, p2) as p2,
               if(p1='', NULL, p3) as p3
        from two
    ),
    four as (
        select nconst, p1 as profession from three where p1 is not NULL
    ),
    five as (
        select nconst, p2 as profession from three where p2 is not NULL
    ),
    six as (
        select nconst, p3 as profession from three where p3 is not NULL
    ),
    seven as (
select * from four
union
select * from five
union
select * from six)
select * from seven order by nconst limit 20;
````

````python
%%sql

create table nconst_profession as 
with one as (select
    nconst,
    primaryProfession,
    substr(primaryProfession, 1, locate(',', primaryProfession)-1) as p1,
    substr(primaryProfession, locate(',', primaryProfession)+1) as remainder
from
    name_basics),
    two as (
        select nconst, primaryProfession, p1,
               substr(remainder, 1, locate(',', remainder)-1) as p2,
               substr(remainder, locate(',', remainder)+1) as p3
        from
            one
    ),
    three as (
        select nconst,
               primaryProfession,
               if(p1='', NULL, p1) as p1,
               if(p2='', NULL, p2) as p2,
               if(p1='', NULL, p3) as p3
        from two
    ),
    four as (
        select nconst, p1 as profession from three where p1 is not NULL
    ),
    five as (
        select nconst, p2 as profession from three where p2 is not NULL
    ),
    six as (
        select nconst, p3 as profession from three where p3 is not NULL
    ),
    seven as (
select * from four
union
select * from five
union
select * from six)
select * from seven;
````

- I would like a generated full course number and a created timestamp.<br><br>
- Most databases support various types of generated columns.

````python
%%sql

use class_example;

drop table if exists mock_course_full;

create table mock_course_full like mock_course;

alter table mock_course_full
    add full_course_number varchar(32) as (concat(dept_code, faculty_code, course_number)) stored;

alter table mock_course_full
    add created_timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP ON UPDATE CURRENT_TIMESTAMP;
    
/* 
    This only works for stored, generated columns.
*/
alter table mock_course_full
    add constraint mock_course_full_pk
        primary key (full_course_number);
        
insert into mock_course_full(dept_code, faculty_code, course_number, course_name)
    select * from mock_course;
````

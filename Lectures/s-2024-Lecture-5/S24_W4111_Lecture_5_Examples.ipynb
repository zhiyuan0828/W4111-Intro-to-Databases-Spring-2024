You also saw an example of my reverse engineering in LucidChart.

| <img src="lucid_chart.jpg"> |
| :---: |
| __Lucidchart Reverse Engineering__ |


````python
%%sql

SELECT 'mysql' dbms,
       t.TABLE_SCHEMA,
       t.TABLE_NAME,
       c.COLUMN_NAME,
       c.ORDINAL_POSITION,
       c.DATA_TYPE,
       c.CHARACTER_MAXIMUM_LENGTH,
       n.CONSTRAINT_TYPE,
       k.REFERENCED_TABLE_SCHEMA,
       k.REFERENCED_TABLE_NAME,
       k.REFERENCED_COLUMN_NAME
FROM INFORMATION_SCHEMA.TABLES t
         LEFT JOIN INFORMATION_SCHEMA.COLUMNS c ON t.TABLE_SCHEMA = c.TABLE_SCHEMA AND t.TABLE_NAME = c.TABLE_NAME
         LEFT JOIN INFORMATION_SCHEMA.KEY_COLUMN_USAGE k
                   ON c.TABLE_SCHEMA = k.TABLE_SCHEMA AND c.TABLE_NAME = k.TABLE_NAME AND c.COLUMN_NAME = k.COLUMN_NAME
         LEFT JOIN INFORMATION_SCHEMA.TABLE_CONSTRAINTS n
                   ON k.CONSTRAINT_SCHEMA = n.CONSTRAINT_SCHEMA AND k.CONSTRAINT_NAME = n.CONSTRAINT_NAME AND
                      k.TABLE_SCHEMA = n.TABLE_SCHEMA AND k.TABLE_NAME = n.TABLE_NAME
WHERE t.TABLE_TYPE = 'BASE TABLE'
  AND t.TABLE_SCHEMA NOT IN ('INFORMATION_SCHEMA', 'mysql', 'performance_schema');
````

# Functions

An example using a bunch of stuff.


Compute the time between an order being placed and being shipped and get the min, max and average by country.

````python
%%sql

with one as (select orderNumber,
                    customerNumber,
                    (select country
                     from customers
                     where customers.customerNumber =
                           orders.customerNumber)    as country,
                    datediff(shippedDate, orderDate) as order_processing_days
             from orders
             where shippedDate is not NULL)
select country,
       count(*)                             as no_of_orders,
       round(avg(order_processing_days), 2) as average_processing_days,
       min(order_processing_days)           as minimum_processing_days,
       max(order_processing_days)           as maximum_processing_days
from one
group by country
order by country;
````

What if we want to recreate the original, uncleaned up table?

````python
%%sql

with one as (select nconst,
                    concat(
                            ifnull(title, ''), ' ',
                            ifnull(first_name, ''), ' ',
                            ifnull(middle_name, ''), ' ',
                            if(nickname is null, '', concat('(', nickname, ') ')),
                            ifnull(last_name, ''), ' ',
                            ifnull(suffix, ''), ' '
                        ) as primaryName
             from name_basics_clean),
     two as (select nconst, primaryName, profession
             from one
                      left join
                  nconst_profession
                  using (nconst)
            order by nconst, profession)
select nconst, primaryName, group_concat(profession) as primaryProfession
from two
group by nconst, primaryName
limit 10;

````
